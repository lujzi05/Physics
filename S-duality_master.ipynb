```python
import numpy as np
import matplotlib.pyplot as plt

# Define S-duality transformation
def tauS(t):
    return -1/t

# Function to plot nested semicircles
def plot_nested_semicircles(N, num_transformations):
    plt.figure(figsize=(8, 8))

    # Start with tau = iN
    tau_values = 1j * np.array([N])

    for _ in range(num_transformations):
        # Create theta values for a semicircle
        theta_values = np.linspace(0, np.pi, 100)
        
        for tau_val in tau_values:
            # Generate the semicircle points
            semicircle_real = np.real(tau_val) + np.cos(theta_values) * np.abs(tau_val)
            semicircle_imag = np.imag(tau_val) + np.sin(theta_values) * np.abs(tau_val)
            plt.plot(semicircle_real, semicircle_imag, 'b', alpha=0.5)

        # Apply S-duality to the current tau values to generate new nested semicircles
        tau_values = tauS(tau_values)
    
    # Plot settings
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.xlabel('Re(τ)')
    plt.ylabel('Im(τ)')
    plt.title('Nested Semicircles under S-Duality')
    plt.grid(True)
    plt.show()

# Parameters
N = 3 # Start with tau = iN, where N is an integer
num_transformations = 3  # Number of S-duality transformations

plot_nested_semicircles(N, num_transformations)
```
